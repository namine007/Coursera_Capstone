**Data Cleaning**

```python
#Remove duplicate entries
venuredf.drop_duplicates(subset='id',keep ='first',inplace=True)

#Extracting information of interest:venure name, categories, locations

filtered_columns = ['name', 'categories','location.address','location.lat','location.lng','location.postalCode']   

venuredf = venuredf.loc[:, filtered_columns] 

# clean column names by keeping only last term 
venuredf.columns = [column.split('.')[-1] for column in venuredf.columns] 


# function that extracts the category of the venue 
def get_category_type(row): 
    try: 
        categories_list = row['categories'] 
    except: 
        categories_list = row['venue.categories']     
    if len(categories_list) == 0: 
        return None 
    else: 
        return categories_list[0]['name'] 

# filter the category for each row 
venuredf['categories'] = venuredf.apply(get_category_type, axis=1) 

#reset index
venuredf = venuredf.reset_index(drop=True)
venuredf
```

**Importing neibourhood/borough information**

```python
#Importing table from wikipedia
wiki = "https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&direction=prev&oldid=926287641"
website_url = urllib.request.urlopen(wiki)

soup = BeautifulSoup(website_url,'lxml')

right_table=soup.find('table',class_='wikitable sortable')

#Parsing wikipedia table into dataframe

#First, creat lists and append the data from wiki html to the list
A=[]
B=[]
C=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        
#Second, create a dataframe and transform the list into dataframe
boroughdf=pd.DataFrame(A,columns=['PostalCode'])
boroughdf['Borough']=B
boroughdf['Neighbourhood']=C
boroughdf
```

**Data Cleaning**

```python
#Remove extra '\n' at the end of cells
boroughdf.replace('\n','',regex=True,inplace=True)

#Remove rows with a borough that is Not assigned.
boroughdf['Borough'].replace('Not assigned', np.nan, inplace=True)
boroughdf.dropna(axis=0, inplace=True)

#Combine rows that share the same postcode & borough
groupeddf=boroughdf.groupby(['PostalCode','Borough'], as_index=False,sort=False).agg(', '.join)

#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
newboroughdf=groupeddf.copy()
for index, row in newboroughdf.iterrows():
    if row[2]=='Not assigned':
        newboroughdf.replace(to_replace=row[2], value =row[1],inplace=True)
newboroughdf
```

**Add Borough/Neighborhood info to venues dataframe**

```python
newvenuredf=venuredf.copy().rename(columns ={'postalCode':'PostalCode'})

newvenuredf['PostalCode'] = newvenuredf['PostalCode'].str[:3]

newvenuredf= newvenuredf.merge(newboroughdf, how = 'left', on='PostalCode')
```

**To narrow down the area where client should open his store, we cluster all Japanese venures into 12 clusters using k-means clustering.**

**Importing libraries for k-means**

```python
import matplotlib.pyplot as plt  

from sklearn.datasets.samples_generator import make_blobs  

%matplotlib inline 
```

```python
k_means = KMeans(init = "k-means++", n_clusters = 12, n_init = 12) 

X = newvenuredf[['lat','lng']].values

k_means.fit(X) 

labels = k_means.labels_ 

print(labels) 

newvenuredf_km = newvenuredf.copy()
newvenuredf_km["Clus_km"] = labels 
```

```python
newvenuredf_km
```

**We count the number of venures in each cluster**

```python
cluster_counts = newvenuredf_km.groupby('Clus_km').count()
cluster_counts
```

```python
cluster_1st = cluster_counts['name'].max()

index_1st = cluster_counts['name'].idxmax()

cluster_2nd = cluster_counts.drop([index_1st])['name'].max()
index_2nd = cluster_counts.drop([index_1st])['name'].idxmax()

print ('Cluster', index_1st, 'has', cluster_1st, 
       'Japanese venures, which is the most, followed by Cluster',
       index_2nd, 'with', cluster_2nd, 'Japanese venures')
```

**We will closely look at 2 clusters with the most venures.**

```python
cluster_most = newvenuredf_km[newvenuredf_km.Clus_km.eq(index_1st)]
cluster_most
```

```python
cluster_2nd_most = newvenuredf_km[newvenuredf_km.Clus_km.eq(index_2nd)]
cluster_2nd_most
```

```python
print('The unique boroughs in ', index_1st,' are: ', pd.unique(cluster_most['Borough'].dropna()))
print('The unique neighbourhoods in ', index_1st,' are: ', 
      pd.unique(cluster_most['Neighbourhood'].dropna()))

print('The unique boroughs in ', index_2nd,' are: ', pd.unique(cluster_2nd_most['Borough'].dropna()))
print('The unique neibourhoods in ', index_2nd,' are: ', pd.unique(cluster_2nd_most['Neighbourhood'].dropna()))
```

**Other than NaN boroughs, these venures are all from Downtown Toronto. Since these k-mean clusters were grouped based on latitude and longitude, we can tell that the venures with NaN boroughs were either in Downtown Toronto or very close to Downtown Toronto.**
